```python
def porter_stemmer(word):
    suffixes = {
        'ational': 'ate',
        'tional': 'tion',
        'enci': 'ence',
        'anci': 'ance',
        'izer': 'ize',
        'bli': 'ble',
        'alli': 'al',
        'entli': 'ent',
        'eli': 'e',
        'ousli': 'ous',
        'ization': 'ize',
        'ation': 'ate',
        'ator': 'ate',
        'alism': 'al',
        'iveness': 'ive',
        'fulness': 'ful',
        'ousness': 'ous',
        'aliti': 'al',
        'iviti': 'ive',
        'biliti': 'ble'
    }

    for suffix in suffixes.keys():
        if word.endswith(suffix):
            return word[:-len(suffix)] + suffixes[suffix]

    return word

text = "Stemming is the process of reducing inflected  words to their word stem, base or root form."
words = text.split()
stemmed_words = [porter_stemmer(word.lower()) for word in words]
stemmed_text = ' '.join(stemmed_words)
print("Original Text:")
print(text)
print("\nStemmed Text:")
print(stemmed_text)
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

nltk.download('punkt')

stemmer = PorterStemmer()

text = "Stemming is the process of reducing inflected  words to their word stem, base or root form."

words = word_tokenize(text)

stemmed_words = [stemmer.stem(word) for word in words]
stemmed_text = ' '.join(stemmed_words)

print("Original Text:")
print(text)
print("\nStemmed Text:")
print(stemmed_text)
```
