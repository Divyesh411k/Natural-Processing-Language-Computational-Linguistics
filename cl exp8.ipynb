```python
import nltk
from nltk.corpus import wordnet
nltk.download('wordnet')
```

```python
# 2. Look up a word using Synsets()
word = "book"
synsets = wordnet.synsets(word)
print("Synsets for the word 'book':", synsets)
```

```python
# 3. Look up a definition of a word found in Synsets
if synsets:
    for synset in synsets:
        print("Definition:", synset.definition())
```

```python
# 4. Look up examples of the words found in Synsets
if synsets:
    for synset in synsets:
        print("Examples:", synset.examples())
```

```python
# 5. Find a set of synonyms that share a common meaning
synonyms = []
if synsets:
    for synset in synsets:
        synonyms.extend(synset.lemmas())
synonyms = set(synonym.name() for synonym in synonyms)
print("Synonyms of 'book':", synonyms)
```

```python
# 6. Find a set of antonyms that do not share a common meaning
antonyms = []
if synsets:
    for synset in synsets:
        for lemma in synset.lemmas():
            if lemma.antonyms():
                antonyms.extend(lemma.antonyms())
antonyms = set(antonym.name() for antonym in antonyms)
print("Antonyms of 'book':", antonyms)
```

```python
# 7. Find a set of hyponyms and hypernyms of a word
hyponyms = set()
hypernyms = set()
if synsets:
    for synset in synsets:
        hyponyms.update(synset.hyponyms())
        hypernyms.update(synset.hypernyms())
hyponyms = set(hyponym.name() for hyponym in hyponyms)
hypernyms = set(hypernym.name() for hypernym in hypernyms)
print("Hyponyms of 'book':", hyponyms)
print("Hypernyms of 'book':", hypernyms)
```
