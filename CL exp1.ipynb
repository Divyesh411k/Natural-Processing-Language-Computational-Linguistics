#1. Tokenization:

Theory:

Breaks down continuous text into smaller units called "tokens," typically words, sentences, or characters.
Aims to create meaningful individual elements for NLP processing.
Can be rule-based .
Applications:

Example:
Input: "This is a paragraph. Another sentence follows."
Output: ["This is a paragraph.", "Another sentence follows."]

#2.Filteration
Selects desired tokens based on specific criteria, reducing noise and redundancy.

Can use regular expressions, linguistic rules, or machine learning models.

a)Character-level: Removes punctuation, symbols, and non-alphanumeric characters.

b)Token-level: Eliminates tokens based on length, case, stop words, or part-of-speech tags.

c)N-gram-level: Filters out n-grams (sequences of tokens) that don't meet semantic or frequency criteria.






#3. Stop Word Removal:

Removes high-frequency, low-information words like articles, prepositions, pronouns, and conjunctions.

Based on the assumption that these words carry little semantic meaning and can clutter analysis.

eg:
Input: "I went to the store to buy apples."

Output: "went store buy apples" (removes articles, prepositions)

#4. Stemming:

Reduces words to their base form (stem) by aggressively chopping suffixes.
Rule-based approach, often using Porter's stemming algorithm.

Can reduce vocabulary size and potentially improve performance in some NLP tasks.

Eg:
Input: "Running, walked, walks, walker"

Output: "run, walk, walk, walk" (stemming using Porter's algorithm)

```python
#Digit removal
after_removal=''
tokens = nltk.word_tokenize(sentence)
no_digits_tokens = [token for token in tokens if not token.isdigit()]
after_removal=' '.join(no_digits_tokens)
after_removal
```

```python
import inflect

def convert_digits_to_words(text):
  p = inflect.engine()
  words = []
  for word in text.split():
    try:
      number = int(word)
      words.append(p.number_to_words(number))
    except ValueError:
      words.append(word)
  return " ".join(words)


converted_text = convert_digits_to_words(sentence)
print(converted_text)  # Output: The price is one hundred twenty-three dollars.
```

```python
#stopword removal
from nltk.corpus import stopwords
en_stops = set(stopwords.words('english'))
all_words = nltk.word_tokenize(sentence)
for word in all_words:
    if word not in en_stops:
        print(word)
```

```python
#stopword_and_punctuation_removal
stop_words = set(stopwords.words('english'))
filtered=[]
word_tokens = word_tokenize(sentence)
filtered = [w for w in word_tokens if  w not in string.punctuation]
filtered
```

```python
#remove white space
no_space = sentence.replace( " ", "")
no_space
```

```python
# stopword_punctuation_number_removal
stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(sentence)
filtered_sentence = []
for w in word_tokens:
    if w not in stop_words and w not in string.punctuation:
      try:
        w = w.replace(',', '')
        if w.isdigit():
           w = num2words(w)
        elif w[:-2].isdigit() and w.endswith(('st', 'nd', 'rd', 'th')):
            w = num2words(w[:-2], to='ordinal')
      except:
          w = ""
          filtered_sentence.append(w)
filtered_sentence
```

```python
# porter_stem_words
porter = PorterStemmer()
por=[]
word_list = word_tokenize(sentence)
for word in word_list:
  a=porter.stem(word)
  por.append(a)
por
```

```python
#lancaster_stem_words
lancaster = LancasterStemmer()
lan=[]
word_list = word_tokenize(sentence)
for word in word_list:
  a=lancaster.stem(word)
  lan.append(a)
lan
```

```python
#lemmatized_output
lemmatizer = WordNetLemmatizer()
word_list = word_tokenize(sentence)
lemmatized_output = [lemmatizer.lemmatize(word) for word in word_list]
lemmatized_output
```
